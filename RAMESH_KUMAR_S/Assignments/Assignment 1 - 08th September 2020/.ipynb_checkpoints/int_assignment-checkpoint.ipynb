```python
#Take one int value between 0 - 256.
#Assign it to two different variables.
#Check the id of both the variables. It should come same. Check why?

#Take one int value either less than -5 or greater than 256.
#Assign it to two different variables.
#Check the id of both the variables. It should come different.Check why?
```

```python
#Solution A:
#Reason for Same ID's
print("Solution A: ")
print("\t Reason for Same ID's")

#Integer variable declaration
num1 = 10
print("Value of variable 'num1' : ", num1)
num2 = 10
print("Value of variable 'num2' : ", num2)

#Id of declared integer variables
print("ID of variable 'num1'    : ", id(num1))
print("ID of variable 'num2'    : ", id(num2))

#Size of declared integer variables
import sys
print("Size of variable 'num1'  : ", sys.getsizeof(num1))
print("Size of variable 'num2'  : ", sys.getsizeof(num2))

#Same ID check
print("Whether Id of variables 'num1' & 'num2' are same ? ", num1 is num2)

#Reason
print("Reason : Python, only creates new reference for already exiting value/object within the range according to their variable datatype. For integer variables whose values within the range -5 to 256, Id's will be same. So, the Id's are same for num1 and num2.")

print()

#Solution B:
print("Solution B: ")
print("\t Reason for Different ID's")

#Integer variable declaration
num3 = -10
print("Value of variable 'num3' : ", num3)
num4 = -10
print("Value of variable 'num4' : ", num4)

#Id of declared integer variables
print("ID of variable 'num3'    : ", id(num3))
print("ID of variable 'num4'    : ", id(num4))

#Size of declared integer variables
import sys
print("Size of variable 'num3'  : ", sys.getsizeof(num3))
print("Size of variable 'num4'  : ", sys.getsizeof(num4))

#Same ID check
print("Whether Id of variables 'num3' & 'num4' are same ? ", num3 is num4)

#Reason
print("Reason : Python, will create new object for the values which lies outside the range according to their variable datatype. For integer variables outside the range -5 to 256, Id's will be different. So, the Id's are different for num3 and num4.")

#Detailed Reason
print()
print("Detailed Reason : ")
print("\t According to concept of Object Reusability and Dynamic Memory Mangement in Python, this language uses the exisitng objects for the variables which are having their values within datatype's range and creates only the reference. And creates new objects for the values which lies outside their datatype's range. Python allocates memory for the variables during the run time dynamically whereas other language allocates memory during the compile time.")
```

```python
#Logical operators
#Observe the output of below code
#Cross check the output manually

print(10 and 20)       #----------------------------------------->Output is 20
print(0 and 20)        #----------------------------------------->Output is 0
print(20 and 0)        #----------------------------------------->Output is 0
print(0 and 0)         #----------------------------------------->Output is 0

print(10 or 20)        #----------------------------------------->Output is 10
print(0 or 20)         #----------------------------------------->Output is 20
print(20 or 0)         #----------------------------------------->Output is 20
print(0 or 0)          #----------------------------------------->Output is 0

print(not 10)          #----------------------------------------->Output is False
print(not 0)           #----------------------------------------->Output is True
```

```python
#Solution :    
print("Solution : ")
print("\t Logical Operator on Integers")

#and Operator on integer values
print("'and' Operator on integer values")
print("Reason : Below operations follows the funtional definition 'A and B returns A if A is False else B'. \n \t According to this 'and' operator evaluates first condition/value, if it is False it will return False without evalauting the second condition/value. If first condition is True, then it will evaluate the second condtion/value then the returns the result.")
print("10 and 20 : ", 10 and 20)
print("0  and 20 : ", 0 and 20)
print("20 and 0  : ", 20 and 0)
print("0  and 0  : ", 0 and 0)

print()
#or Operator on integer values
print("'or' Operator on integer values")
print("Reason : Below operations follows the funtional definition 'A or B returns A if A is True else B'. \n \t According to this 'or' operator evaluates first condition/value, if it is True it will return True without evalauting the second condition/value. If first condition is False, then it will evaluate the second condtion/value then the returns the result.")
print("10  or 20 : ", 10 or 20)
print("0   or 20 : ", 0 or 20)
print("20  or 0  : ", 20 or 0)
print("0   or 0  : ", 0 or 0)

print()
#not Operator on integer values
print("'not' Operator on integer values")
print("Reason : Below operations follows the funtional definition 'not A returns False if A is True else True'. \n \t According to this 'not' operator evaluates the condition/value, if it results True it will return False and if it results False, it will return True.")
print("not 10    : ", not 10)
print("not 0     : ", not 0)
```

```python
#Solution :    
print("Solution : ")
print("\t Bitwise Operator on Integers")

#Bitwise and Operator on integer values
print("Bitwise 'and' Operator")
print("Bitwise 'and' Operator on integer values 10 & 20         : 10 & 20 = ", 10 & 20)
print("Reason : Bitwise 'and' returns '1' if both input results '1', else it will return '0'")
print("Manual Check : ")
print("Binary value of 10                                       : ", (bin(10)))
print("Binary value of 20                                       : ", (bin(20)))
print("Bitwise 'and' of 10 & 20 in Binary value                 :  00000")                         
print("Decimal value of 0000                                    : ", (0b00000))
print("Hence Verified")
print()

#Bitwise or Operator on integer values
print("Bitwise 'or' Operator")
print("Bitwise 'or' Operator on integer values 10 & 20         : 10 | 20 = ", 10 | 20)
print("Reason : Bitwise 'or' returns '0' if both input results '0', else it will return '1'")
print("Manual Check : ")
print("Binary value of 10                                       : ", (bin(10)))
print("Binary value of 20                                       : ", (bin(20)))
print("Bitwise 'or' of 10 | 20 in Binary value                  :  11110")                         
print("Decimal value of 11110                                   : ", (0b11110))
print("Hence Verified")
print()

#Bitwise xor Operator on integer values
print("Bitwise 'xor' Operator")
print("Bitwise 'xor' Operator on integer values 10 & 20         : 10 | 20 = ", 10 | 20)
print("Reason : Bitwise 'xor' returns '0' if both inputs are same, else it will return '1'")
print("Manual Check : ")
print("Binary value of 10                                       : ", (bin(10)))
print("Binary value of 20                                       : ", (bin(20)))
print("Bitwise 'xor' of 10 ^ 20 in Binary value                 :  11110")                         
print("Decimal value of 11110                                   : ", (0b11110))
print("Hence Verified")
print()

#Bitwise negation Operator on integer value
print("Bitwise 'negation' Operator")
print("Bitwise 'negation' Operator on integer value 10          : ~10 = ", ~10)
print("Reason : Bitwise 'negation' follows the functional logic ~(value)=-(value+1)")
print("Manual Check : ")                    
print("-(value + 1)                                             : ", -(10 + 1))
print("Hence Verified")
print()

#Bitwise left shift Operator on integer values
print("Bitwise 'left shift' Operator")
print("Bitwise 'left shift' Operator on integer values 10 << 2  : 10 << 2 = ", 10 << 2)
print("Reason : Bitwise 'left shift' converts 10 into binary value then shifts each binary digits two steps left and appends two zeros at right end ")
print("Manual Check : ")
print("Binary value of 10                                       : ", (bin(10)))
print("Binary value of 10 in 32 bits                            :  00000000000000000000000000001010")
print('''After shifting binary digits two steps left 
       and appending two zeros at right end              :  00000000000000000000000000101000''')                        
print("Binary value of 00000000000000000000000000101000         : ", (0b00000000000000000000000000101000))
print("Hence Verified")
print()

#Bitwise right shift Operator on integer values
print("Bitwise 'right shift' Operator")
print("Bitwise 'right shift' Operator on integer values 10 >> 2 : 10 >> 2 = ", 10 >> 2)
print("Reason : Bitwise 'right shift' converts 10 into binary value then shifts each binary digits two steps riht and appends two zeros at left end ")
print("Manual Check : ")
print("Binary value of 10                                       : ", (bin(10)))
print("Binary value of 10 in 32 bits                            :  00000000000000000000000000001010")
print('''After shifting binary digits two steps left 
       and appending two zeros at right end              :  00000000000000000000000000000010''')                        
print("Binary value of 00000000000000000000000000000010         : ", (0b00000000000000000000000000000010))
print("Hence Verified")
print()
```

```python
#Solution :  
print("Solution : ")
print("\t Identity Operator on Integers")
print("Solution A : ")

#Integer variable declaration
print()
a = 10
print("Value of variable 'a' : ", a)
b = 10
print("Value of variable 'b' : ", b)

print("'a is b' will result as 'True'")
print("'a is not b' will result as 'False'")
print("Reason : Python, only creates new reference for already exiting value/object within the range according to their variable datatype. For integer variables within the range -5 to 256, Id's will be same. Here, the Id's of a and b are same so 'a is b' results True and 'a is not b' results False.")

#Id of declared integer variables
print("ID of variable 'a'    : ", id(a))
print("ID of variable 'b'    : ", id(b))
print("a is b                : ", a is b)
print("a is not b            : ", a is not b)

print()
print("Solution B : ")

#Integer variable declaration
print()
a = 1000
print("Value of variable 'a' : ", a)
b = 1000
print("Value of variable 'b' : ", b)

print("'a is b' will result as 'False'")
print("'a is not b' will result as 'True'")
print("Reason : Python, will create new object for the values which lies outside the range according to their variable datatype. For integer variables outside the range -5 to 256, Id's will be different. Here, the Id's of a and b are different so 'a is b' results False and 'a is not b' results True.")

#Id of declared integer variables
print("ID of variable 'a'    : ", id(a))
print("ID of variable 'b'    : ", id(b))
print("a is b                : ", a is b)
print("a is not b            : ", a is not b)
```
